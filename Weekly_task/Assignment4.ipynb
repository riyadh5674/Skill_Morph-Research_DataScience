# **Task 1(Matplotlib)**

## **Bar chart**

```python
import matplotlib.pyplot as plt

# Count membership types
membership_counts = df['abonoment_type'].value_counts()

# Plot bar chart
plt.figure(figsize=(8,5))
membership_counts.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title("Gym Membership Types")
plt.xlabel("Membership Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

## **Histogram**

```python
# Histogram of average time spent in gym
plt.figure(figsize=(8,5))
plt.hist(df['avg_time_in_gym'], bins=10, color='lightgreen', edgecolor='black')

plt.title("Distribution of Average Time in Gym")
plt.xlabel("Average Time (minutes)")
plt.ylabel("Frequency")
plt.show()
```

## **Scatterplot**

```python
# Scatter plot of Age vs Visits per week
plt.figure(figsize=(8,5))
plt.scatter(df['Age'], df['visit_per_week'], alpha=0.6, color='coral', edgecolors='black')

plt.title("Age vs. Visits per Week")
plt.xlabel("Age")
plt.ylabel("Visits per Week")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

```

# **Task 2(Seaborn)**

## **Box plot**

```python
# Box plot: Gender vs Age
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='gender', y='Age', palette='Set2')

plt.title("Box Plot of Age by Gender")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```

## **Count plot**

```python
# Count plot: Days per week
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='days_per_week', palette='pastel', edgecolor='black')

plt.title("Count of Gym Visits per Week")
plt.xlabel("Days per Week")
plt.ylabel("Number of Members")
plt.show()
```

## **Bar plot**

```python
# Bar plot: Average visits per week by membership type
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='abonoment_type', y='visit_per_week', palette='Set2', ci=None)

plt.title("Average Visits per Week by Membership Type")
plt.xlabel("Membership Type")
plt.ylabel("Average Visits per Week")
plt.xticks(rotation=45)
plt.show()
```
