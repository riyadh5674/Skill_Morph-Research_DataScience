```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 30, .01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='Day', ylabel='Practice > 1 > Theory',
       title='My learning curve')
ax.grid()

# fig.savefig("test.png")
plt.show()
```
